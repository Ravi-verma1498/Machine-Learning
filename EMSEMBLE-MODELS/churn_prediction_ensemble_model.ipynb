```python
# Feature enginering
df['balance_stability'] = df[['current_month_balance', 'previous_month_balance']].std(axis=1)
df['average_balance_ratio'] = df['average_monthly_balance_prevQ'] / (df['average_monthly_balance_prevQ2'] + 1)
df['transaction_amount_ratio'] = df['current_month_credit'] / (df['current_month_debit'] + 1)  # Avoid division by zero
df['transaction_frequency'] = df['days_since_last_transaction'] / df['vintage']


df
```

## Features Descriptions

*cutomer id *: unique identifier   

vintage : duration for which customer associated with the bank 
*age *: the age of the customer
gender : gender of customer
dependents: the no of dependents of customer has
occupation : occupation type(salaried ,self-employed,retired)
customer_nw_category : net worth category(low,medium,high)
days_since_last_transaction:  no of days customer made last transaction
current_balance : ammout in customer account
previous_month_end_balance : accunt balanance at the end of last month
average_monthly_balance_prevQ: avg monthly balance of customer in last quarter
average_monthly_balance_prevQ2: avg monthly balance of customer in quarter                                     before last quarter
current_month_credit: total amount of debit transaction(withdrawl)
previous_month_debit: total debit transactions made in the previous month
current_month_balance:current_month_balance
previous_month_balance:balance in the previous month. By comparing this with                         the current month balance, one can track trends or                             anomalies in the account.    
churn : target variable (1 for churned and 0 for not churned)

## dealing with outliers

```python
def replace_outliers_with_mean(df):
    for column in df.select_dtypes(include=[np.number]).columns:
        
        if column != 'churn':
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3-Q1
            lower_bound = Q1=1.5*IQR
            upper_bound = Q3+1.5*IQR
            mean_value = df[column].mean()
        else:
            continue
        df[column]=np.where((df[column]<lower_bound)|(df[column]>upper_bound),mean_value,df[column])
        return df

df = replace_outliers_with_mean(df)
```

```python
scaler = StandardScaler()
features = ['customer_id',
 'vintage',
 'dependents',
 'age',
 'days_since_last_transaction',
 'current_balance',
 'previous_month_end_balance',
 'average_monthly_balance_prevQ',
 'average_monthly_balance_prevQ2',
 'current_month_credit',
 'previous_month_credit',
 'current_month_debit',
 'previous_month_debit',
 'current_month_balance',
 'previous_month_balance',
 'balance_stability',
 'average_balance_ratio',
 'transaction_amount_ratio',
 'transaction_frequency']

df[features] = scaler.fit_transform(df[features])
df
```

```python
df_encoded = pd.get_dummies(df,columns = ['gender','occupation','customer_nw_category'],
                            drop_first = True)
df_encoded
```

```python
x = df_encoded.drop(['churn','days_since_last_transaction'],axis = 1)
y = df['churn']

xtrain ,xtest,ytrain,ytest = train_test_split(x,y,test_size =0.2,random_state = 42)
xtest
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
n_estimators = 100, # no of trees
max_depth = 10,
# min_samples_split = 5,
# min_samples_split=2,
random_state = 42) 
rf.fit(xtrain,ytrain)
ypred_rf = rf.predict(xtest)

rf_accuracy = accuracy_score(ytest,ypred_rf)
rf_classification_report =classification_report(ytest,ypred_rf) 
rf_confusion_mat = confusion_matrix(ytest,ypred_rf)

print("rf_accuracy: ",rf_accuracy)
print("rf_classification_report: ",rf_classification_report)
print("rf_confusion_mat: " ,rf_confusion_mat)

```

```python
# !pip install -c conda-forge xgboost
import xgboost as xgb
from xgboost import XGBClassifier
xgb_model =XGBClassifier(n_estimators = 100,learning_rate = 0.1,random_state = 42) 
xgb_model.fit(xtrain,ytrain)

ypred_xgb = xgb_model.predict(xtest)

xgb_accuracy = accuracy_score(ytest,ypred_xgb)
xgb_classification_report =classification_report(ytest,ypred_xgb) 
xgb_confusion_mat = confusion_matrix(ytest,ypred_xgb)
print("xgb_accuracy: ",xgb_accuracy)
print("xgb_classification_report: ",xgb_classification_report)
print("xgb_confusion_mat: " ,xgb_confusion_mat)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

base_models= [
     ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('xgb', XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42))
]
meta_model = LogisticRegression()
stacking_model = StackingClassifier(estimators = base_models,
                                   final_estimator = meta_model)

stacking_model.fit(xtrain,ytrain)
ypred_st =stacking_model.predict(xtest)
st_accuracy = accuracy_score(ytest,ypred_st)
st_classification_report =classification_report(ytest,ypred_st) 
st_confusion_mat = confusion_matrix(ytest,ypred_st)
print("st_accuracy: ",st_accuracy)
print("st_classification_report: ",st_classification_report)
print("st_confusion_mat: " ,st_confusion_mat)

```

```python
from sklearn.metrics import roc_auc_score

# Calculate AUC-ROC for each model
rf_auc = roc_auc_score(ytest, rf.predict_proba(xtest)[:, 1])
xgb_auc = roc_auc_score(ytest, xgb_model.predict_proba(xtest)[:, 1])
st_auc = roc_auc_score(ytest, stacking_model.predict_proba(xtest)[:, 1])

print(f"Random Forest AUC-ROC: {rf_auc}")
print(f"XGBoost AUC-ROC: {xgb_auc}")
print(f"Stacking AUC-ROC: {st_auc}")
```
